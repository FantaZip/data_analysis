```python
#괄호로 된 주소 제거
def remove_brackets(address):
    return address.split('(')[0]

df['도로명주소'] = df['도로명주소'].apply(remove_brackets)
```

```python
# 콤마 뒤에 있는 거 지우기
def split(address):
    split_address = address.split(',')
    first_sentence = split_address[0].strip()
    return first_sentence
    
df['도로명주소'] = df['도로명주소'].apply(split)
```

```python
def remove_apartment_and_join(address):
    # 문자열을 공백을 기준으로 split
    words = address.split()

    # '아파트'가 포함된 단어를 찾아 제외
    words = [word for word in words if '백화점' not in word]

    # 나머지 단어들을 공백으로 join
    cleaned_address = ' '.join(words)

    return cleaned_address

df['지번주소'] = df['지번주소'].apply(remove_apartment_and_join)
```

```python
import re
def remove(address):
    words = address.split()
    result1 = words[0:5]
    result = [words[i] + ' ' + words[i + 1] for i in range(len(words) - 1) if re.match('.*동$', words[i])]
    joined_result = ' '.join(result1)
    return joined_result
df['지번주소'] = df['지번주소'].apply(remove)
```
