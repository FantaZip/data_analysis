```python
grouped_data = merged.groupby("level")[["not_sexCrime_h12030w", "sexCrime_h12030w"]].mean()

# grouped_data.plot(kind='bar', figsize=(10, 6))
grouped_data.plot(kind='bar')
plt.title("Average Values by Level")
plt.xlabel("Level")
plt.ylabel("Average Values")
plt.legend(title="Features", bbox_to_anchor=(1, 1))  # 범례 위치 조정
plt.show()
```

```python
from scipy import stats

test_names = ["not_sexCrime_h12030w", "sexCrime_h12030w"]

safe_scores = merged[merged['level'] == 'safe'][["not_sexCrime_h12030w", "sexCrime_h12030w"]]
danger_scores = merged[merged['level'] == 'danger'][["not_sexCrime_h12030w", "sexCrime_h12030w"]]

for test_name in test_names:
    _, p_value_levene = stats.levene(safe_scores[test_name], danger_scores[test_name])
    if p_value_levene > 0.05:
        print(f"{test_name} p-value: {p_value_levene}, 등분산 가정 만족")
    else:
        print(f"{test_name} p-value: {p_value_levene}, 이분산 가정 만족")
```

```python
t_statistic, p_value = stats.ttest_ind(
    a=safe_scores,
    b=danger_scores,
    alternative="less",    # safe < danger
    equal_var=False
)
print(f"p-value: {p_value}")
print(f"귀무 가설 기각: {p_value < 0.05}")
```
